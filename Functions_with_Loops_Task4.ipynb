PROBLEM 1: Mobile Recharge Validation **System**

```python
print("=" * 60)
print("PROBLEM 1: MOBILE RECHARGE VALIDATION SYSTEM")
print("=" * 60)

def validate_recharge(amount, plan):
    """
    Validates mobile recharge request based on amount and plan.
    Returns True if valid, False otherwise.
    """
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Error: Minimum recharge amount is ₹50")
        return False

    if plan not in valid_plans:
        print(f"Error: Invalid plan. Valid plans: {valid_plans}")
        return False

    return True

# Main program with retry loop
while True:
    try:
        amt = int(input("Enter recharge amount (₹): "))
        pln = int(input("Enter plan (199/299/399/599): "))

        if validate_recharge(amt, pln):
            print(f"Recharge successful! Amount: ₹{amt}, Plan: ₹{pln}")
            break
        else:
            retry = input("Try again? (yes/no): ").lower()
            if retry != "yes":
                print("Recharge cancelled.")
                break
    except ValueError:
        print("Please enter valid numbers.")
        retry = input("Try again? (yes/no): ").lower()
        if retry != "yes":
            break

print()


```

PROBLEM 2: Inventory Reorder Alert **System**

```python
print("=" * 60)
print("PROBLEM 2: INVENTORY REORDER ALERT SYSTEM")
print("=" * 60)

def check_inventory(inventory):
    """
    Checks stock levels and prints reorder alerts.
    Threshold: 15 units
    """
    print("\nInventory Status:")
    print("-" * 30)

    for product, stock in inventory.items():
        if stock < 15:
            status = "REORDER ALERT"
        else:
            status = "Stock OK"

        print(f"{product}: {stock} units - {status}")

# Sample inventory
products = {
    "Laptop": 12,
    "Mouse": 45,
    "Keyboard": 8,
    "Monitor": 20,
    "USB Cable": 5
}

check_inventory(products)
print()
```

PROBLEM 3: Student Result Processing **System**

```python
print("=" * 60)
print("PROBLEM 3: STUDENT RESULT PROCESSING SYSTEM")
print("=" * 60)

def calculate_result(marks):
    """
    Calculates average marks and returns Pass/Fail status.
    Pass criteria: Average >= 50
    """
    total = 0
    for mark in marks:
        total += mark

    average = total / len(marks)

    print(f"Marks: {marks}")
    print(f"Average: {average:.2f}")

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

student_marks = [65, 78, 52, 45, 80]
result = calculate_result(student_marks)
print(f"Result: {result}")
print()
```

 PROBLEM 4: Cab Fare Estimator with Retry **Option**

```python
print("=" * 60)
print("PROBLEM 4: CAB FARE ESTIMATOR")
print("=" * 60)

def calculate_fare(distance, is_peak_hour):
    """
    Calculates cab fare with base fare, per km charge,
    and peak hour surcharge.
    """
    base_fare = 50
    per_km = 12

    fare = base_fare + (distance * per_km)

    if is_peak_hour:
        fare = fare * 1.25  # 25% extra
        print("Peak hour surcharge applied (25%)")

    return fare

# Main program with retry loop
while True:
    try:
        dist = float(input("Enter distance (km): "))
        peak = input("Is it peak hour? (yes/no): ").lower()

        is_peak = (peak == "yes")
        total_fare = calculate_fare(dist, is_peak)

        print(f"Total fare: ₹{total_fare:.2f}")

        retry = input("Calculate another fare? (yes/no): ").lower()
        if retry != "yes":
            print("Thank you for using our service!")
            break

    except ValueError:
        print("Please enter valid distance.")
        retry = input("Try again? (yes/no): ").lower()
        if retry != "yes":
            break

print()
```

PROBLEM 5: Employee Attendance Eligibility **Checker**

```python
print("=" * 60)
print("PROBLEM 5: ATTENDANCE ELIGIBILITY CHECKER")
print("=" * 60)

def check_attendance_eligibility(attendance_list):
    """
    Calculates attendance percentage and returns eligibility.
    Eligible if attendance >= 75%
    """
    present_count = 0
    total_days = len(attendance_list)

    for status in attendance_list:
        if status == "P":
            present_count += 1

    percentage = (present_count / total_days) * 100

    print(f"Total days: {total_days}")
    print(f"Present: {present_count}")
    print(f"Attendance: {percentage:.2f}%")

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance = ["P", "P", "A", "P", "P", "P", "A", "P", "P", "P"]
eligibility = check_attendance_eligibility(attendance)
print(f"Status: {eligibility}")
print()
```

PROBLEM 6: Password Strength **Checker**

```python
print("=" * 60)
print("PROBLEM 6: PASSWORD STRENGTH CHECKER")
print("=" * 60)

def check_password_strength(password):
    """
    Checks password strength based on length, digits,
    and special characters.
    """
    # Check minimum length
    if len(password) < 8:
        return "Weak: Password must be at least 8 characters"

    # Check for digit
    has_digit = False
    for char in password:
        if char.isdigit():
            has_digit = True
            break

    if not has_digit:
        return "Weak: Must contain at least one digit"

    # Check for special character
    special_chars = "@#$"
    has_special = False
    for char in password:
        if char in special_chars:
            has_special = True
            break

    if not has_special:
        return "Weak: Must contain at least one special character (@#$)"

    return "Strong: Password meets all criteria"

# Test passwords
test_passwords = ["hello", "hello123", "hello@world", "Hello1@world"]

for pwd in test_passwords:
    result = check_password_strength(pwd)
    print(f"Password: '{pwd}' -> {result}")

print("=" * 60)
```
