# **Task 1 - User Login Check**

```python

username = "admin"
password = "1234"

input_user = input("Enter username: ")
input_pass = input("Enter password: ")

if input_user == username and input_pass == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

# Task 2 - Pass/Fail **Analyzer**

```python

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print(f"Total students: {len(marks)}")
print(f"Passed: {pass_count}")
print(f"Failed: {fail_count}")
```

# **Task 3 - Simple Data Cleaner**

```python

names = [" Alice ", "bob", " CHARLIE "]
cleaned = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned.append(cleaned_name)

print("Original:", names)
print("Cleaned:", cleaned)
```

# Task 4 - Message Length **Analyzer**

```python

messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    length = len(msg)
    if length > 10:
        print(f"'{msg}' → {length} chars [FLAGGED: Too long]")
    else:
        print(f"'{msg}' → {length} chars [OK]")
```

# Task 5 - Error Message **Detector**

```python

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print(f"Total logs: {len(logs)}")
print(f"Error count: {error_count}")
```
