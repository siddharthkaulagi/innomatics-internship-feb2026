# Task 3: Data Processing and **Analysis**

PROBLEM 1: Employee Performance Bonus Eligibility

```python
print("=" * 60)
print("PROBLEM 1: EMPLOYEE PERFORMANCE BONUS ELIGIBILITY")
print("=" * 60)

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

highest_score = 0
for score in employees.values():
    if score > highest_score:
        highest_score = score


top_performers = []
for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {highest_score})")
print()
```

 PROBLEM 2: Search Query Keyword Analysis

```python
print("=" * 60)
print("PROBLEM 2: SEARCH QUERY KEYWORD ANALYSIS")
print("=" * 60)

query = "Buy mobile phone buy phone online"


cleaned = query.lower()
punctuation = ".,!?;:"
for char in punctuation:
    cleaned = cleaned.replace(char, "")

words = cleaned.split()


word_count = {}
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1


duplicates = {}
for word, count in word_count.items():
    if count > 1:
        duplicates[word] = count

print(duplicates)
print()
```

PROBLEM 3: Sensor Data Validation

```python
print("=" * 60)
print("PROBLEM 3: SENSOR DATA VALIDATION")
print("=" * 60)

sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for hour, reading in enumerate(sensor_readings):
    if reading % 2 == 0:  # Check if even
        valid_readings.append((hour, reading))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
print()
```

PROBLEM 4: Email Domain Usage Analysis

```python
print("=" * 60)
print("PROBLEM 4: EMAIL DOMAIN USAGE ANALYSIS")
print("=" * 60)

emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Extract domains and count
domain_count = {}
for email in emails:
    domain = email.split("@")[1]
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Calculate percentages
total = len(emails)
for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {percentage:.0f}%")
print()
```

 PROBLEM 5: Sales Spike Detection

```python
print("=" * 60)
print("PROBLEM 5: SALES SPIKE DETECTION")
print("=" * 60)

sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
total_sales = sum(sales)
average = total_sales / len(sales)

# Find spikes (30% above average)
threshold = average * 1.30

print(f"Average sales: {average:.2f}")
print(f"Spike threshold (30% above): {threshold:.2f}")
print("Spike days:")

for day, sale in enumerate(sales, start=1):
    if sale > threshold:
        print(f"Day {day}: {sale}")
print()
```

PROBLEM 6: Duplicate User ID Detection

```python
print("=" * 60)
print("PROBLEM 6: DUPLICATE USER ID DETECTION")
print("=" * 60)

user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Count occurrences
id_count = {}
for user_id in user_ids:
    if user_id in id_count:
        id_count[user_id] += 1
    else:
        id_count[user_id] = 1

# Display only duplicates
for user_id, count in id_count.items():
    if count > 1:
        print(f"{user_id} → {count} times")

print("=" * 60)
```
