PROBLEM 1: Smart Light **Controller**

```python
print("=" * 60)
print("PROBLEM 1: SMART LIGHT CONTROLLER")
print("=" * 60)

class SmartLight:
    """
    Class to control a smart light with ON/OFF functionality.
    """
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.name} is {self.status}")

# Test
light = SmartLight("Bedroom Light")
light.turn_on()
light.display_status()
print()
```

PROBLEM 2: Employee ID Card **System**

```python
print("=" * 60)
print("PROBLEM 2: EMPLOYEE ID CARD SYSTEM")
print("=" * 60)

class EmployeeIDCard:
    """
    Class to store and display employee ID card details.
    """
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

# Test
employee = EmployeeIDCard("Rahul", "EMP102", "IT")
employee.display_card()
print()
```

PROBLEM 3: Mobile Contact **Record**

```python
print("=" * 60)
print("PROBLEM 3: MOBILE CONTACT RECORD")
print("=" * 60)

class Contact:
    """
    Class to store and display contact information.
    """
    def __init__(self, name, phone_number):
        self.name = name
        self.phone_number = phone_number

    def save_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone_number}")

# Test
contact = Contact("Anita", "9876543210")
contact.save_contact()
print()
```

PROBLEM 4: Product Price Tag **Generator**

```python
print("=" * 60)
print("PROBLEM 4: PRODUCT PRICE TAG GENERATOR")
print("=" * 60)

class ProductPriceTag:
    """
    Class to store product details and print formatted price tag.
    """
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price

    def generate_tag(self):
        print(f"Product: {self.product_name}")
        print(f"Price: ₹{self.price}")

# Test
product = ProductPriceTag("Headphones", 2499)
product.generate_tag()
print()
```

PROBLEM 5: Movie Rating Display **System**

```python
print("=" * 60)
print("PROBLEM 5: MOVIE RATING DISPLAY SYSTEM")
print("=" * 60)

class MovieRating:
    """
    Class to store movie details and display rating.
    """
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating

    def display_details(self):
        print(f"Movie: {self.movie_name}")
        print(f"Rating: {self.rating} / 5")

# Test
movie = MovieRating("Inception", 4.8)
movie.display_details()
print()
```

 PROBLEM 6: Delivery Address **Manager**

```python
print("=" * 60)
print("PROBLEM 6: DELIVERY ADDRESS MANAGER")
print("=" * 60)

class DeliveryAddress:
    """
    Class to store and print delivery details.
    """
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def print_delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")

# Test
delivery = DeliveryAddress("Suman", "Hyderabad")
delivery.print_delivery_details()
print("=" * 60)
```
