 PROBLEM 1: Social Media – Post Engagement **Analyzer**

```python
print("=" * 60)
print("PROBLEM 1: POST ENGAGEMENT ANALYZER")
print("=" * 60)

def analyze_engagement(likes_list):
    """
    Calculates total likes and determines if post is viral.
    Viral threshold: 1000+ likes
    """
    total_likes = 0
    for like in likes_list:
        total_likes += like

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    return total_likes, status

# Test
daily_likes = [120, 340, 210, 180, 200]
total, post_status = analyze_engagement(daily_likes)

print(f"Total Likes: {total}")
print(f"Post Status: {post_status}")
print()
```

PROBLEM 3: Agriculture – Rainfall Adequacy **Checker**

```python
print("=" * 60)
print("PROBLEM 3: RAINFALL ADEQUACY CHECKER")
print("=" * 60)

def check_rainfall_adequacy(rainfall_readings, required_level):
    """
    Calculates average rainfall and checks adequacy.
    """
    total_rainfall = 0
    for reading in rainfall_readings:
        total_rainfall += reading

    average = total_rainfall / len(rainfall_readings)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    return average, status

# Test
seasonal_rainfall = [65, 80, 70, 75, 70]
required = 70
avg_rainfall, rain_status = check_rainfall_adequacy(seasonal_rainfall, required)

print(f"Average Rainfall: {avg_rainfall}")
print(f"Rainfall Status: {rain_status}")
print()

```

PROBLEM 4: Social Media – Duplicate Account **Detection**

```python
print("=" * 60)
print("PROBLEM 4: DUPLICATE ACCOUNT DETECTION")
print("=" * 60)

def detect_duplicate_accounts(usernames):
    """
    Detects if duplicate usernames exist in system.
    Uses set to find unique vs total count.
    """
    unique_usernames = set(usernames)

    if len(unique_usernames) < len(usernames):
        return "Yes"
    else:
        return "No"

# Test
user_accounts = ["alice123", "bob456", "alice123", "charlie789"]
duplicate_found = detect_duplicate_accounts(user_accounts)

print(f"Duplicate Accounts Found: {duplicate_found}")
print()
```

PROBLEM 6: Agriculture – Premium Crop Price **Filter**

```python
print("=" * 60)
print("PROBLEM 6: PREMIUM CROP PRICE FILTER")
print("=" * 60)

def filter_premium_crops(prices, threshold=2000):
    """
    Filters crop prices above premium threshold.
    Default threshold: ₹2000
    """
    premium_crops = []
    for price in prices:
        if price > threshold:
            premium_crops.append(price)

    return premium_crops

# Test
crop_prices = [1500, 2500, 1800, 3200, 2100, 1900]
premium_list = filter_premium_crops(crop_prices)

print(f"Premium Crops: {premium_list}")
print()
```

 PROBLEM 7: System Monitoring – Application Health **Checker**

```python
print("=" * 60)
print("PROBLEM 7: APPLICATION HEALTH CHECKER")
print("=" * 60)

def check_application_health(error_count):
    """
    Evaluates system health based on error count.
    Healthy: 0 errors
    Minor Issues: 1-5 errors
    Critical Issues: >5 errors
    """
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    return status

# Test
errors = 7
health_status = check_application_health(errors)

print(f"Error Count: {errors}")
print(f"System Status: {health_status}")
print()
```

```python
print("=" * 60)
print("PROBLEM 8: DAILY TRANSACTION LIMIT CHECKER")
print("=" * 60)

def check_transaction_limit(transaction_amount, daily_limit=50000):
    """
    Validates daily transaction against limit.
    Default limit: ₹50,000
    """
    if transaction_amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    return status

# Test
transaction = 60000
transaction_status = check_transaction_limit(transaction)

print(f"Transaction Amount: {transaction}")
print(f"Transaction Status: {transaction_status}")
print()
```

PROBLEM 9: E-Learning – Student Attendance **Eligibility**

```python
print("=" * 60)
print("PROBLEM 9: STUDENT ATTENDANCE ELIGIBILITY SYSTEM")
print("=" * 60)

def check_attendance_eligibility(attendance_list):
    """
    Calculates attendance percentage and checks exam eligibility.
    Eligibility threshold: 75%
    """
    present_count = 0
    for status in attendance_list:
        if status == "P":
            present_count += 1

    total_days = len(attendance_list)
    percentage = (present_count / total_days) * 100

    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    return percentage, eligibility

# Test
attendance_record = ["P", "P", "A", "P", "P", "P", "A", "P", "P", "P"]
attendance_pct, exam_eligibility = check_attendance_eligibility(attendance_record)

print(f"Attendance Percentage: {attendance_pct}")
print(f"Exam Eligibility: {exam_eligibility}")
print()
```

PROBLEM 10: Smart Electricity Bill **Analyzer**

```python
print("=" * 60)
print("PROBLEM 10: SMART ELECTRICITY BILL ANALYZER")
print("=" * 60)

def calculate_electricity_bill(units_consumed):
    """
    Calculates electricity bill based on slab rates.
    Classifies usage level based on total bill amount.

    Slabs:
    First 100 units: ₹3/unit
    Next 100 units: ₹5/unit
    Above 200 units: ₹7/unit

    Classification:
    < ₹500: Low Usage
    ₹500-₹1500: Moderate Usage
    > ₹1500: High Usage
    """
    # Calculate bill based on slabs
    if units_consumed <= 100:
        bill = units_consumed * 3
    elif units_consumed <= 200:
        bill = (100 * 3) + ((units_consumed - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units_consumed - 200) * 7)

    # Classify usage
    if bill < 500:
        usage_level = "Low Usage"
    elif bill <= 1500:
        usage_level = "Moderate Usage"
    else:
        usage_level = "High Usage"

    return bill, usage_level

# Test
units = 250
total_bill, usage_status = calculate_electricity_bill(units)

print(f"Units Consumed: {units}")
print(f"Total Bill: ₹{total_bill}")
print(f"Usage Classification: {usage_status}")
print("=" * 60)
```
