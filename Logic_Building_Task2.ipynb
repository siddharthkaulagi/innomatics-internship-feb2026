PROBLEM 1: Unique Words in a Sentence

```python
print("=" * 50)
print("PROBLEM 1: UNIQUE WORDS IN A SENTENCE")
print("=" * 50)

sentence = input("Enter sentence: ")  # python is easy and python is powerful

# Split into words and convert to set for uniqueness
words = sentence.split()
unique_words = set(words)

print(f"Unique words count: {len(unique_words)}")
print(f"Unique words: {unique_words}")
print()
```

PROBLEM 2: Highest Salary from Employee Data

```python
print("=" * 50)
print("PROBLEM 2: HIGHEST SALARY FINDER")
print("=" * 50)

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with maximum salary
max_salary = 0
top_employee = ""

for name, salary in employees.items():
    if salary > max_salary:
        max_salary = salary
        top_employee = name

print(f"Highest Salary: {top_employee} - {max_salary}")
print()
```

PROBLEM 3: Find Maximum and Minimum Values

```python
print("=" * 50)
print("PROBLEM 3: MAX AND MIN VALUES")
print("=" * 50)

numbers = [45, 22, 89, 10, 66]

# Initialize with first element
max_val = numbers[0]
min_val = numbers[0]

# Compare with remaining elements
for num in numbers[1:]:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print(f"List: {numbers}")
print(f"Max: {max_val}")
print(f"Min: {min_val}")
print()
```

PROBLEM 4: Count Products Above Price Threshold

```python
print("=" * 50)
print("PROBLEM 4: PRODUCTS ABOVE THRESHOLD")
print("=" * 50)

prices = [450, 1200, 899, 1500, 300]
threshold = 1000

count = 0
for price in prices:
    if price > threshold:
        count += 1

print(f"Products above {threshold}: {count}")
print()
```

PROBLEM 5: Calculate Attendance Percentage

```python
print("=" * 50)
print("PROBLEM 5: ATTENDANCE PERCENTAGE")
print("=" * 50)

attendance = ["P", "P", "A", "P", "P"]

present_count = 0
for status in attendance:
    if status == "P":
        present_count += 1

percentage = (present_count / len(attendance)) * 100
print(f"Attendance Percentage: {percentage}")
print()
```

PROBLEM 7: Count Character Frequency

```python
print("=" * 50)
print("PROBLEM 7: CHARACTER FREQUENCY")
print("=" * 50)

text = "python"
freq = {}

for char in text:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1

print(freq)
print()
```
