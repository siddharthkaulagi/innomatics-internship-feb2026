 PROBLEM 1: Smart Parking Lot Management **System**

```python
print("=" * 60)
print("PROBLEM 1: SMART PARKING LOT MANAGEMENT SYSTEM")
print("=" * 60)

def manage_parking(capacity, logs):
    """
    Manages parking lot based on entry/exit logs.
    Calculates current occupancy and checks capacity.
    """
    current_parked = 0
    peak_usage = 0

    for log in logs:
        if log == "IN":
            current_parked += 1
        elif log == "OUT":
            current_parked -= 1

        # Track peak usage
        if current_parked > peak_usage:
            peak_usage = current_parked

    # Check capacity alert
    if current_parked > capacity:
        alert = "CAPACITY EXCEEDED!"
    elif current_parked == capacity:
        alert = "Full"
    else:
        alert = "Available"

    return current_parked, peak_usage, alert

# Test
parking_capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

parked, peak, status = manage_parking(parking_capacity, vehicle_logs)

print(f"Currently Parked Vehicles: {parked}")
print(f"Peak Usage: {peak}")
print(f"Parking Status: {status}")
if parked > parking_capacity:
    print(f"ALERT: Parking exceeds capacity by {parked - parking_capacity} vehicles!")
print()
```

PROBLEM 2: Online Food Delivery Time **Estimator**

```python
print("=" * 60)
print("PROBLEM 2: ONLINE FOOD DELIVERY TIME ESTIMATOR")
print("=" * 60)

def estimate_delivery_time(distance, traffic, weather):
    """
    Estimates delivery time based on distance, traffic and weather.
    Base speed: 20 km/h
    Applies multipliers for traffic and weather conditions.
    """
    # Base time calculation (20 km/h average speed)
    base_time = (distance / 20) * 60  # Convert to minutes

    # Traffic multipliers
    traffic_multipliers = {
        "Low": 1.0,
        "Medium": 1.3,
        "High": 1.6
    }

    # Weather multipliers
    weather_multipliers = {
        "Clear": 1.0,
        "Cloudy": 1.1,
        "Rainy": 1.4,
        "Stormy": 1.8
    }

    # Apply multipliers
    traffic_factor = traffic_multipliers.get(traffic, 1.0)
    weather_factor = weather_multipliers.get(weather, 1.0)

    estimated_time = base_time * traffic_factor * weather_factor

    return int(estimated_time)

# Test
dist = 8
traffic_level = "High"
weather_condition = "Rainy"

delivery_time = estimate_delivery_time(dist, traffic_level, weather_condition)

print(f"Distance: {dist} km")
print(f"Traffic Level: {traffic_level}")
print(f"Weather: {weather_condition}")
print(f"Estimated Delivery Time: {delivery_time} minutes")
print()
```

 PROBLEM 3: Movie Theatre Seat Occupancy **Analyzer**

```python
print("=" * 60)
print("PROBLEM 3: MOVIE THEATRE SEAT OCCUPANCY ANALYZER")
print("=" * 60)

def analyze_occupancy(total_seats, booked_seats):
    """
    Analyzes seat booking data and determines show status.
    booked_seats is a list where 1 = booked, 0 = available
    """
    booked_count = 0
    for seat in booked_seats:
        booked_count += seat

    occupancy_percentage = (booked_count / total_seats) * 100

    # Determine show status
    if occupancy_percentage == 100:
        show_status = "Housefull"
    elif occupancy_percentage >= 90:
        show_status = "Almost Full - Open Additional Show Recommended"
    elif occupancy_percentage >= 75:
        show_status = "Almost Full"
    elif occupancy_percentage >= 50:
        show_status = "Good Occupancy"
    else:
        show_status = "Low Occupancy"

    return occupancy_percentage, booked_count, show_status

# Test
total = 200
booked = [1] * 150  # 150 seats booked

occupancy, booked_total, status = analyze_occupancy(total, booked)

print(f"Total Seats: {total}")
print(f"Booked Seats: {booked_total}")
print(f"Occupancy: {occupancy:.0f}%")
print(f"Show Status: {status}")
print()
```

PROBLEM 4: Cloud Server Load Classification **System**

```python
print("=" * 60)
print("PROBLEM 4: CLOUD SERVER LOAD CLASSIFICATION SYSTEM")
print("=" * 60)

def classify_server_load(cpu_readings):
    """
    Classifies server load based on CPU usage readings.
    Categories: Normal (<50%), Warning (50-80%), Critical (>80%)
    """
    # Calculate average
    total = 0
    for reading in cpu_readings:
        total += reading

    average_cpu = total / len(cpu_readings)

    # Classify load
    if average_cpu < 50:
        status = "Normal"
    elif average_cpu <= 80:
        status = "Warning"
    else:
        status = "Critical"

    return average_cpu, status

# Test
cpu_data = [45, 60, 70, 85, 90]

avg_load, server_status = classify_server_load(cpu_data)

print(f"CPU Readings: {cpu_data}")
print(f"Average CPU Load: {avg_load:.0f}%")
print(f"Server Status: {server_status}")

if server_status == "Critical":
    print("ALERT: Immediate attention required!")
elif server_status == "Warning":
    print("CAUTION: Monitor closely for spikes.")
print()
```

 PROBLEM 5: Smart Classroom Resource Usage **Monitor**

```python
print("=" * 60)
print("PROBLEM 5: SMART CLASSROOM RESOURCE USAGE MONITOR")
print("=" * 60)

def monitor_resource_usage(usage_data, threshold=8):
    """
    Tracks classroom resource usage and identifies overuse.
    Default threshold: 8 hours
    """
    overused = []
    total_usage = 0

    for resource, hours in usage_data.items():
        total_usage += hours
        if hours > threshold:
            overused.append(resource)

    # Determine energy alert
    if len(overused) > 0:
        energy_alert = "Yes"
    else:
        energy_alert = "No"

    return overused, total_usage, energy_alert

# Test
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

overused_resources, total, alert = monitor_resource_usage(resource_usage)

print(f"Resource Usage: {resource_usage}")
print(f"Total Usage Hours: {total}")

if overused_resources:
    print(f"Overused Resources: {', '.join(overused_resources)}")
else:
    print("Overused Resources: None")

print(f"Energy Alert: {alert}")

if alert == "Yes":
    print("RECOMMENDATION: Optimize AC usage to save energy.")
print()
```

PROBLEM 6: Online Event Registration Capacity **Controller**

```python
print("=" * 60)
print("PROBLEM 6: ONLINE EVENT REGISTRATION CAPACITY CONTROLLER")
print("=" * 60)

def manage_event_registration(capacity, registrations):
    """
    Manages event registrations with waitlist functionality.
    Prevents overbooking and tracks waitlisted users.
    """
    confirmed = 0
    waitlisted = 0

    for i in range(registrations):
        if i < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    # Determine registration status
    if registrations >= capacity:
        status = "Closed"
    else:
        status = "Open"

    return confirmed, waitlisted, status

# Test
event_capacity = 100
total_registrations = 105

confirmed_count, waitlist_count, reg_status = manage_event_registration(event_capacity, total_registrations)

print(f"Event Capacity: {event_capacity}")
print(f"Total Registrations: {total_registrations}")
print(f"Confirmed Registrations: {confirmed_count}")
print(f"Waitlisted Users: {waitlist_count}")
print(f"Registration Status: {reg_status}")

if waitlist_count > 0:
    print(f"NOTE: {waitlist_count} users added to waitlist.")
print("=" * 60)
```
